# Import Libraries

```python
import pandas as pd
import numpy as np
```

# Reading Data

```python
data = pd.ExcelFile('export-registrasi(1) - passinggrade.xlsx')

all_data = dict()

all_data['ketm'] = pd.read_excel(data, 'ketm')
all_data['kondisi-tertentu'] = pd.read_excel(data, 'kondisi-tertentu')
all_data['perpindahan'] = pd.read_excel(data, 'perpindahan')
all_data['prestasi pil_1'] = pd.read_excel(data, 'prestasi pil_1')
all_data['persiapan-kelas-industri'] = pd.read_excel(data, 'persiapan-kelas-industri')
all_data['jarak'] = pd.read_excel(data, 'jarak')
# all_data['rapor'] = pd.read_excel(data, 'rapor') # Ini datanya kosong, jadi gaperlu
all_data['prestasi pil_2'] = pd.read_excel(data, 'prestasi pil_2')
all_data['persiapan-kelas-industri'] = pd.read_excel(data, 'persiapan-kelas-industri')
all_data['Sheet1'] = pd.read_excel(data, 'Sheet1')
```

# Cleaning Data
Setelah di analisis, ternyata semua dataset pada file excel di mulai pada index ke 3 atau baris ke 4. **Kecuali Sheet1**

Jadi yang perlu kita lakukan adalah set column name dataset dari baris ke 4, dan slicing data yang diperlukan saja

```python
for sheet in all_data.keys():    
    all_data[sheet].columns = all_data[sheet].iloc[2] # Extract Column Name dari baris ke-3
    all_data[sheet] = all_data[sheet].iloc[3:, :] # Slicing Data 
    
    #     Lakukan Pengecualian pada Sheet1
    if sheet == "Sheet1":
        all_data[sheet] = all_data[sheet].set_index('No').rename_axis(None, axis=1) # Set kolom No. sebagai Index, dan menghapus nama index
    else:
        all_data[sheet] = all_data[sheet].set_index('No.').rename_axis(None, axis=1) # Set kolom No. sebagai Index, dan menghapus nama index
    
    
    
    
```

# Cleaning Data Sheet1

```python
all_data['Sheet1'].head()
```

Dari dataset **Sheet1** di atas kita dapat menyimpulkan satu record data melingkupi 4 baris, serta pada kolom skor terdapat beberapa data, yaitu skor kejuaraan, skor uji kompensi, dan skor akhir. Data yang memiliki value pada seluruh record yang ada pada kolom skor hanyalah data skor kejuaraan. jadi sisanya tidak perlu kita ambil

```python
sheet_1 = all_data['Sheet1'].loc[all_data['Sheet1'].index.notna()] # Ambil row pertama setiap record
```

```python
sheet_1.info()
```

Kesimpulan Informasi data:
Kita perlu menghapus 4 kolom **nan**. karena kolom tersebut tidak berguna

```python
sheet_1 = sheet_1.loc[:, sheet_1.columns[sheet_1.columns.notna()]] # Hapus Kolom NaN
```

```python
all_data['Sheet1'] = sheet_1
```
